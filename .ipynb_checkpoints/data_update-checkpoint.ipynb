```python
import psycopg2

conn = psycopg2.connect(host="localhost", port = 5432, database="Eric_Health", user="postgres", password="Keaton0506!!")

cur = conn.cursor()
cur.execute("SELECT date FROM days")

dates = []
res = cur.fetchall()
for r in res:
    dates.append(r[0])
    
for d in dates:
    print(d)
    exercise_dist_durs = 0
    workout_dur = 0
    cur.execute("SELECT * FROM exercise_distance WHERE date = '{}'".format(d))
    res = cur.fetchall()
    for r in res:
        exercise_dist_durs += r[4]
    cur.execute("SELECT * FROM workouts WHERE date = '{}'".format(d))
    res = cur.fetchall()
    for r in res:
        workout_dur += r[3]
    ex_dur = exercise_dist_durs + workout_dur
    cur.execute("UPDATE days SET exercise_duration = {} WHERE date = '{}'".format(int(ex_dur), d))
    conn.commit()
    print("", int(ex_dur))
cur.close()
conn.close()
```
