```python
import psycopg2
import csv
import os
import statistics

data_path = r"Z:\Dev\plotly-charts\health_data"

conn = psycopg2.connect(host="localhost", port = 5432, database="Eric_Health", user="postgres", password="Keaton0506!!")

cur = conn.cursor()
cur.execute("SELECT date, mood FROM days ORDER BY date")

mood_map = {
    "happy": 5,
    "anxious": 2,
    "stressed": 2,
    "overwhelmed": 2,
    "good": 4,
    "disappointed": 2,
    "busy": 3,
    "meh": 3,
    "great": 6,
    "content": 5,
    "energetic": 4,
    "productive": 4,
    "claustrophobic": 2,
    "angry": 1,
    "annoyed": 2,
    "bad": 1,
    "sad": 1,
    "accomplished": 4,
    "frustrated": 2,
    "lethargic": 2,
    "relaxed": 5
}

res = cur.fetchall()
for r in res:
    print(r[0])
    scores = []
    no_space = r[1].replace(" ", "")
    mood_list = no_space.split(",")
    for n in mood_list:
        print(n, mood_map[n])
        scores.append(mood_map[n])
    print(statistics.mean(scores))
    cur.execute("UPDATE days SET mood_int = {} WHERE date = '{}'".format(statistics.mean(scores), r[0]))
    conn.commit()

cur.close()
conn.close()
```
