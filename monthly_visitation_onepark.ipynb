```python
import pandas as pd
import plotly.graph_objects as go

# df = pd.read_csv(r"Z:\Dev\plotly-charts\NPS_VIsitation_CSVs\ROMO.csv")
# df = pd.read_csv(r"Z:\Dev\plotly-charts\NPS_VIsitation_CSVs\JOTR.csv")
# df = pd.read_csv(r"Z:\Dev\plotly-charts\NPS_VIsitation_CSVs\ZION.csv")
df = pd.read_csv(r"Z:\Dev\plotly-charts\NPS_VIsitation_CSVs\GLAC.csv")
```

```python
years = df['Year'].tolist()
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

yr = []
for y in years:
    yr.append(str(y))
zs = []
for m in months:
    temp_list = []
    temp_list.append(df[m].tolist())
    num_list = []
    for t in temp_list:
        for i in t:
            try:
                num_list.append(int(str(i).replace(",", "")))
            except:
                num_list.append(None)
    zs.append(num_list)

fig = go.Figure(data=go.Heatmap(
        z=[zs[11],zs[10],zs[9],zs[8],zs[7],zs[6],zs[5],zs[4],zs[3],zs[2],zs[1],zs[0]],
        y=['DEC','NOV','OCT','SEP','AUG','JUL','JUN','MAY','APR','MAR','FEB','JAN'],
        x=yr,
        colorscale='sunsetdark'))

fig.update_layout(
    
    font=dict(family='Courier New, monospace', size=10, color='#7f7f7f')
)

fig.show()
```
