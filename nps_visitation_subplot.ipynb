```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(
    rows=2, cols=2,
    column_widths=[0.5, 0.5],
    specs=[[{"type": "heatmap"},{"type": "heatmap"}],[{"type": "heatmap"},{"type": "heatmap"}]],
    subplot_titles=("Glacier","Rocky Mountain", "Zion", "Joshua Tree")
)

#GLAC
glac_df = pd.read_csv(r"Z:\Dev\plotly-charts\NPS_VIsitation_CSVs\GLAC.csv")
years = glac_df['Year'].tolist()
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

yr = []
for y in years:
    yr.append(str(y))
zs = []
for m in months:
    temp_list = []
    temp_list.append(glac_df[m].tolist())
    num_list = []
    for t in temp_list:
        for i in t:
            try:
                num_list.append(int(str(i).replace(",", "")))
            except:
                num_list.append(None)
    zs.append(num_list)

fig.add_trace(go.Heatmap(
        z=[zs[11],zs[10],zs[9],zs[8],zs[7],zs[6],zs[5],zs[4],zs[3],zs[2],zs[1],zs[0]],
        y=['DEC','NOV','OCT','SEP','AUG','JUL','JUN','MAY','APR','MAR','FEB','JAN'],
        x=yr,
        colorscale='Viridis'),row=1,col=1)

# ROMO
romo_df = pd.read_csv(r"Z:\Dev\plotly-charts\NPS_VIsitation_CSVs\ROMO.csv")
years = romo_df['Year'].tolist()
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

yr = []
for y in years:
    yr.append(str(y))
zs = []
for m in months:
    temp_list = []
    temp_list.append(romo_df[m].tolist())
    num_list = []
    for t in temp_list:
        for i in t:
            try:
                num_list.append(int(str(i).replace(",", "")))
            except:
                num_list.append(None)
    zs.append(num_list)

fig.add_trace(go.Heatmap(
        z=[zs[11],zs[10],zs[9],zs[8],zs[7],zs[6],zs[5],zs[4],zs[3],zs[2],zs[1],zs[0]],
        y=['DEC','NOV','OCT','SEP','AUG','JUL','JUN','MAY','APR','MAR','FEB','JAN'],
        x=yr,
        colorscale='Viridis'),row=1,col=2)

# ZION
zion_df = pd.read_csv(r"Z:\Dev\plotly-charts\NPS_VIsitation_CSVs\ZION.csv")
years = zion_df['Year'].tolist()
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

yr = []
for y in years:
    yr.append(str(y))
zs = []
for m in months:
    temp_list = []
    temp_list.append(zion_df[m].tolist())
    num_list = []
    for t in temp_list:
        for i in t:
            try:
                num_list.append(int(str(i).replace(",", "")))
            except:
                num_list.append(None)
    zs.append(num_list)

fig.add_trace(go.Heatmap(
        z=[zs[11],zs[10],zs[9],zs[8],zs[7],zs[6],zs[5],zs[4],zs[3],zs[2],zs[1],zs[0]],
        y=['DEC','NOV','OCT','SEP','AUG','JUL','JUN','MAY','APR','MAR','FEB','JAN'],
        x=yr,
        colorscale='Viridis'),row=2,col=1)

# JOTR
jotr_df = pd.read_csv(r"Z:\Dev\plotly-charts\NPS_VIsitation_CSVs\JOTR.csv")
years = jotr_df['Year'].tolist()
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

yr = []
for y in years:
    yr.append(str(y))
zs = []
for m in months:
    temp_list = []
    temp_list.append(jotr_df[m].tolist())
    num_list = []
    for t in temp_list:
        for i in t:
            try:
                num_list.append(int(str(i).replace(",", "")))
            except:
                num_list.append(None)
    zs.append(num_list)

fig.add_trace(go.Heatmap(
        z=[zs[11],zs[10],zs[9],zs[8],zs[7],zs[6],zs[5],zs[4],zs[3],zs[2],zs[1],zs[0]],
        y=['DEC','NOV','OCT','SEP','AUG','JUL','JUN','MAY','APR','MAR','FEB','JAN'],
        x=yr,
        colorscale='Viridis'),row=2,col=2)

fig.update_layout(font=dict(family='Courier New, monospace', size=10, color='#7f7f7f'))
fig.show()
```
